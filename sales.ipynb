```python
# Importing Libraries

import pandas as pd
```

```python
# reading file 
data = pd.read_csv(r"C:\Users\Amit\Desktop\Dataset\sales_data.csv");
```

```python
data
```

```python
# Creating a dataframe for this data 
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# correcting the datatype of order date
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
# extracting day,month and year data from the date column
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
df['Day'] = df['Order Date'].dt.day
```

```python
df.head(10)
```

```python
#checking the null or blank values
df.isna().sum()
```

```python
# Calculating the margin percentage for each data
df['Profit'] = df['Price Each'] - df['Cost price']

df['Margin %'] = (df['Profit']/df['Price Each']) * 100
df
```

```python
df.columns
```

```python
# Extract the state from the address using regular expressions
df['State'] = df['Purchase Address'].str.extract(r', ([A-Z]{2})')
df.head() 
```

```python
# Checking out different states
states = df['State'].unique()
states
```

```python
 df['State'].nunique()
```

```python
# Checking out different unique products
products = df['Product'].unique()
products
```

```python
df['Product'].nunique()
```

```python
# renaming the column's name  
df.rename(columns={"catégorie":"Category"} , inplace=True)
```

```python
# Extract the city from the address using regular expressions
df['City'] = df['Purchase Address'].str.extract(r', ([a-zA-Z\s]+),')
df.head() 
```

```python
# Different cities
df['City'].unique()
```

```python
df['City'].nunique()
```

```python
new_df = df.drop(columns=['Order Date','Product_ean', 'Category','Purchase Address','margin'])
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
new_df.columns
```

```python
# finding out the correlation between these columns
correlation_matrix = new_df.corr()
```

```python
correlation_matrix
```

```python
#Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(correlation_matrix , cmap='coolwarm'  , linewidths=.30)
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.show()
```

```python
total_profit_by_state = new_df.groupby(['State']).Profit.sum()
total = total_profit_by_state.astype(int).sort_values(ascending=False)
total
```

```python
profit_state_df = pd.DataFrame(total)
profit_state_df
```

```python
plt.figure(figsize=(15,8))
plt.bar(profit_state_df.index, profit_state_df['Profit'])

for i, value in enumerate(profit_state_df['Profit']):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.xlabel('States')
plt.ylabel('Profit')
plt.title("Total number of profit per state")
plt.xticks(rotation=45)
plt.show()
```

```python
# mapping the month name

month_name = {1:'Jan',
             2:'Feb',
             3:'Mar',
             4:'Apr',
             5:'May',
             6:'Jun',
             7:'Jul',
             8:'Aug',
             9:'Sep',
             10:'Oct',
             11:'Nov',
             12:'Dec'};

new_df['Month'] = new_df['Month'].map(month_name)
```

```python
new_df
```

```python
profit_in_every_month = new_df.groupby(['Month']).Profit.sum()
```

```python
profit_in_every_month_total = profit_in_every_month.astype(int).sort_values(ascending=False)
```

```python
month_profit_df = pd.DataFrame(profit_in_every_month_total)
```

```python
month_profit_df
```

```python
plt.figure(figsize=(15,10))
plt.bar(month_profit_df.index , month_profit_df['Profit'])

for i, value in enumerate(month_profit_df['Profit']):
    plt.text(i, value, str(value), ha='center', va='bottom')
plt.xlabel("Months")
plt.ylabel("Profit")
plt.title("Total Profit in each month ")
plt.xticks(rotation=45)
plt.show()
```

```python
popular_product = new_df.groupby(['Product']).Profit.sum()
```

```python
popular_product_total = popular_product.astype(int).sort_values(ascending=False) 
```

```python
popular_product_total
```

```python
popular_product_total_df = pd.DataFrame(popular_product_total)
```

```python
popular_product_total_df
```

```python
plt.figure(figsize=(20,9))
bars = plt.barh(popular_product_total_df.index , popular_product_total_df['Profit'])

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.0f}', 
             va='center', ha='left')
    

plt.xlabel("Profits")

plt.title("Profit of each products")
plt.show()
```
